```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
flight_log = pd.read_csv('flight_log.csv')
print(flight_log.head())
print(flight_log.shape)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(flight_log['time'], flight_log['imu_x'], label='Altitude')
plt.plot(flight_log['time'], flight_log['imu_y'], label='Altitude')
plt.plot(flight_log['time'], flight_log['imu_z'], label='Altitude')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(flight_log['time'], flight_log['gyro_x'], label='Altitude')
plt.plot(flight_log['time'], flight_log['gyro_y'], label='Altitude')
plt.plot(flight_log['time'], flight_log['gyro_z'], label='Altitude')
```
